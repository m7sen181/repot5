```python
# drop columns with high nan and columns with no importance
df = df.drop(['Incident_Number', 'How_Long_Delayed',	'Boro', 'Busbreakdown_ID', 'School_Year', 'Bus_No', 'Occurred_On', 'Created_On', 'Informed_On', 'Last_Updated_On','School_Age_or_PreK','Route_Number'] , axis= 1)
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.info()
```

```python
catgorical = ['Reason', 'Run_Type', 'Bus_Company_Name']

hotEncode = OneHotEncoder(sparse_output=False)

categorical_encode = hotEncode.fit_transform(df[catgorical])

# Convert to DataFrame
categorical_encoded_df = pd.DataFrame(categorical_encode, columns=hotEncode.get_feature_names_out(catgorical))

df_encoded = pd.concat([df, categorical_encoded_df], axis=1)

df_encoded = df_encoded.drop(catgorical, axis=1)

df = df_encoded
```

```python
label_encoder = LabelEncoder()

features_encoded = ['Schools_Serviced', 'Has_Contractor_Notified_Schools', 'Has_Contractor_Notified_Parents', 'Have_You_Alerted_OPT', 'Breakdown_or_Running_Late']

for column in features_encoded:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

## Split the Dataset
Next, split the dataset into training, validation, and testing sets.

```python
X = df.drop('Breakdown_or_Running_Late', axis=1)
y = df['Breakdown_or_Running_Late']
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train, y_test = train_test_split(X,y,train_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Building the ANN Model
In this section, define the architecture of the ANN by specifying the number of layers, neurons, and activation functions.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
model = Sequential()
```

```python
model.add(Dense(units = 8 ,input_dim=X_train.shape[1], kernel_initializer='uniform',activation='relu'))
```

```python
model.add(Dense(units = 8 , activation = 'relu'))
model.add(Dense(units = 8 , activation = 'relu'))
```

```python
model.add(Dense(units = 1, activation = 'sigmoid'))
```

## Compile the Model
Compile the ANN model by defining the optimizer, loss function, and evaluation metrics.

```python
model.compile(optimizer='adam', loss='binary_crossentropy',metrics = ['accuracy'])
```

## Training the Model
Train the ANN model using the training data.

```python
model.fit(X_train, y_train, batch_size=140, epochs=22, validation_split=0.2)
```

## Evaluate the Model
Evaluate the performance of the model on the test set.

```python
model.evaluate(X_test, y_test)
```

## Make Predictions
Use the trained model to make predictions on new or unseen data.
